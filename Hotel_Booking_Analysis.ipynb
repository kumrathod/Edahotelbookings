# **Importing The Libraries**

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Loading The Dataset**

```
df = pd.read_csv ("/content/Hotel Bookings.csv")
```

```
# changing the data type of reservation status date
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'] )
```

```
df.describe()
```

```
df.isnull().sum()
```

```
df.drop(['company','agent'],axis =1,inplace = True)
df.dropna(inplace = True)
```

```
df.isnull().sum()
```

```
df = df[df['adr']<5000] #removing outliers
```

# **Data Analysis and Visualization**

```
df['hotel'].value_counts()
```

-> We have got only 2 types of hotels namely City Hotel and Resort Hotel

```
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
print(cancelled_perc)
```

```
plt.figure(figsize = (5,4))
plt.title("Reservation Status")
plt.bar(['Not canceled','Canceled'],df['is_canceled'].value_counts())
plt.show()
```

```
df['adults'].groupby(df['hotel']).mean()
```

```
df['children'].groupby(df['hotel']).mean()
```

-> We conclude that adults and childrens prefer Resort Hotel over City Hotel

```
# adding a column for canceled and non canceled values (1,0) with canceled and not_canceled
df['cancelation'] = df.is_canceled.replace(to_replace= [1,0] , value= ['canceled', 'not_canceled'])
```

```
df_canceled = df['cancelation']
df_canceled.value_counts().plot(kind = 'pie',autopct = "%.0f%%")
plt.title("cancelation plot for hotel booking customers")
```

```
cancelation_by_hotel= df[df['cancelation']== 'canceled'].groupby(['hotel']).size()
cancelation_by_hotel
```

-> Here we conclude that cancelation of city hotel bookings are more compared to resort hotel

```
percentage_of_cancelation_CityHotel = cancelation_by_hotel['City Hotel']/df['hotel'].value_counts() ['City Hotel'] * 100
percentage_of_cancelation_CityHotel
```

```
percentage_of_cancelation_ResortHotel = cancelation_by_hotel['Resort Hotel']/df['hotel'].value_counts() ['Resort Hotel'] * 100
percentage_of_cancelation_ResortHotel
```

```
df_hotel_by_year= df[['arrival_date_year', 'hotel']].value_counts().groupby('arrival_date_year').sum()
```

```
plt.rcParams['figure.figsize'] = (8,5)
df_hotel_by_year.plot(kind = 'bar',color = ['blue','orange','red'])
plt.title("Year-Wise Bookings")
plt.xlabel("Arrival Year")
plt.ylabel("Count of Arrival")
plt.show()
```

-> Most Arrivals are observed in the Year 2016 which is nearly 2.5x then year 2015 and nearly 0.5x then year 2016

```

plt.rcParams['figure.figsize'] = (8,5)
sns.countplot(data = df, x ='arrival_date_year',hue = 'hotel')
plt.title("Count of Arrival per year each hotel")
plt.xlabel("Arrival Year")
plt.ylabel("Arrival Count")
plt.show()
```

```
plt.rcParams['figure.figsize']= (15,5)
sns.countplot(data = df , x = 'arrival_date_month', hue = 'hotel', order = ['January','February','March','April','May','June','July','August','September','October','November','December'])
plt.title("Count of arrival per month each hotel")
plt.xlabel("Arrival Month")
plt.ylabel("Count")
plt.show()
```

```
plt.rcParams['figure.figsize'] = (20,5)
sns.countplot(data = df,x = 'arrival_date_day_of_month',hue = 'hotel')
plt.title("Count of arrival per day each hotel")
plt.xlabel("Arrival Date")
plt.ylabel("Count")
plt.show()
```

-> month of august saw major arrivals in both hotels

-> and there are no such major arrivals in particular days

```
plt.rcParams['figure.figsize'] = (20,5)
sns.countplot(data =df, x ='stays_in_weekend_nights')
plt.title("Count of stays on weekend nights")
plt.xlabel("Stays on weekend nights")
plt.ylabel("Count")
plt.show()
```

```
plt.rcParams['figure.figsize'] = (20,5)
sns.countplot(data = df, x = 'stays_in_week_nights')
plt.title("Count of stays on week nights")
plt.xlabel("Stays on week nights")
plt.ylabel("Count")
plt.show()
```

-> From above observations we can culminate that arrival of guests is more on week nights then weekend nights

```
#identifying adults count from the data
df_adults = df[['hotel','adults']][df['adults']>0].groupby(['hotel']).count()
df_adults
```

```
plt.rcParams['figure.figsize'] = (20,5)
sns.countplot(data = df , x = 'adults', hue = 'hotel').set_title('Counts of adults')
plt.show()
```

```
#identifying children count from the data
df_children= df[['hotel','children']][df['children']>0].groupby(['hotel']).count()
df_children
```

```
plt.rcParams['figure.figsize']= (20,5)
sns.countplot(data = df, x= 'children', hue= 'hotel').set_title('Count of children')
```

```
#identifying babies count from the data
df_babies= df [['hotel','babies']] [df['babies']>0].groupby(['hotel']).count()
df_babies
```

```
plt.rcParams['figure.figsize']= (20,5)
sns.countplot(data = df, x= 'babies', hue= 'hotel').set_title('Count of babies')
```

-> Majorty of the guests come in couple as we can see from count of adults plot, both the hotels acheive peak at 2.0 and spoecifically City Resort could be tagged as a "Duo/Couple-Friendly Hotel"

-> Guests bringing children with them have more preference to City Hotel. Hence, City Hotel could also be tagged as a "Children-Friendly Hotel"

-> Guests bringing babies with them have no such preference over Hotels, but a bit more towards Resort Hotel. So, Resort Hotel could be tagged as "Babies-Friendly Hotel"

```
#guests who have not canceled their booking will only be our case analysis data
#Grouping by countries of the guests
df_country_guests= df[['country', 'hotel']][df['cancelation'] == 'not_canceled'].groupby(['country']).count()
df_country_guests_top10= df_country_guests.sort_values( by= 'hotel', ascending=False).head (10)
```

```
df_country_guests_top10
```

```
plt.rcParams['figure.figsize']= (15,5)
df_country_guests_top10.plot(kind= 'bar' , fontsize=20, color= ['green'])
plt.ylabel('visitors count', fontsize=20)
plt.xlabel('country of visitors', fontsize=20)
plt.title('Top 10 countries by visitors', fontsize=20)
```
